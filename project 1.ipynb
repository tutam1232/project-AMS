## Mô tả các hàm và ý tưởng thực hiện: 
### Hàm Gauss_elimination(A):
- Input: ma trận A là ma trận mở rộng của hệ phương trình Ax=b
- Output: ma trận có dạng bậc thang có được từ ma trận A

Ý tưởng thực hiện:
1. Với cột đầu tiên:
- Nếu chứa toàn bộ số 0 &rarr; sang cột kế tiếp
- Chứa ít nhất một số hạng khác 0 &rarr; sang bước 2
Việc kiểm tra trên được thực đánh dấu bằng biến flag trong hàm (nếu flag True thì sang bước 2, false thì sang cột kế tiếp)
2. Nếu phần tử đầu cột = 0 &rarr; hoán vị dòng này với dòng bên dưới có phần tử trong cột khác 0
3. Nhân dòng 1 cho 1/a (a là số hạng đầu dòng)
4. Cộng dòng bên dưới cho một số thích hợp cho dòng đầu để biến đổi = 0
5. Lặp lại các bước trên với ma trận đã che dòng đã làm

### Hàm back_substitution(A)
- Input: ma trận có dạng bậc thang thu được từ ma trận mở rộng của hệ phương trình Ax=b
- Output: nghiệm của hệ phương trình (trường hợp nghiệm duy nhất/vô số nghiệm) hoặc thông báo hệ phương trình vô nghiệm.

Ý tưởng thực hiện:
1. Đầu tiên kiểm tra trường hợp vô nghiệm của hệ phương trình: nếu có bất kì dòng nào mà có tất cả các phần tử đều bằng 0 (trừ phần tử cuối dòng) và phần tử cuối dòng khác 0 &rarr; thông báo hệ phương trình vô nghiệm, kết thúc hàm
2. Kiểm tra trường hợp vô số nghiệm nếu hệ phương trình không vô nghiệm: thông báo hệ phương trình vô số nghiệm và in nghiệm tổng quát ra màn hình, sau đó kết thúc hàm
3. Giải nghiệm duy nhất nếu hệ phương trình không vô nghiệm và cũng không vô số nghiệm:
- B1: Khởi tạo vector nghiem để lưu các nghiệm của hệ phương trình
- B2: Đưa nghiệm đầu tiên vào nghiem, là phần tử thuộc hàng cuối cùng, cột cuối cùng của ma trận input
- B3: Xét từ dòng kế cuối trở lên tới dòng đầu tiên:
    + Lấy phần tử cuối của dòng đang xét - (từng phần tử còn lại * nghiệm tương ứng đã tìm được)
    + Đưa kết quả trên vào biến nghiem
- B4: Sau khi thực hiện xong B3, vector nghiem sẽ lưu nghiệm duy nhất của phương trình
