```python
#Họ tên: Đàm Tử Tâm
#MSSV: 21120551
#Lớp: 21_2

#Hàm deepcopy_matrix: copy ma trận A ra một bản mới
def deepcopy_matrix(A):
    C=[]
    for i in range(len(A)):
        temp=[]
        for j in range(len(A[i])):
            temp.append(A[i][j]*1)
        C.append(temp)
    return C

#Hàm print_matrix: in ma trận chuyển trạng thái ra màn hình cùng ý nghĩa của các dòng, các cột
def print_matrix(A,state):
    if A==None or len(A)!=len(state):
        return
    else:
        print(''.ljust(15),end='')
        print('|',end='')
        
        for i in range(len(state)):
            if i==0:
                print(''.ljust(5),end='')
            print(state[i].ljust(15),end='')
        print('')
        
        print(''.ljust(15*(len(state)+1)+5,'-'))
        
        for i in range(len(A)):
            print(state[i].ljust(15),end='')
            print('|',end='')
            print(''.ljust(5),end='')
            for j in range(len(A[i])):
                print(str(A[i][j]).ljust(15),end='')
            print('')
        
#Hàm multi_matrix: hàm nhân hai ma trận
def multi_matrix(A,B):
    if(len(A[0])!=len(B)):
        print("Khong the nhan hai ma tran")
        return None
    C=[]
    temp=[]
    ans=0
    for i in range(len(A)):
        temp=[]
        for j in range(len(B[0])):
            ans=0
            for k in range(len(A[0])):
                ans=ans+A[i][k] * B[k][j]
            temp.append(round(ans,3))
        C.append(temp)
    return C

#Hàm markov_chain: hàm tìm vector trạng thái sau "day" ngày để dự báo thời tiết tương ứng sau "day" ngày
def markov_chain(A, v, day):
    C=deepcopy_matrix(v)        
    if day == 0:
        return C    
   
    tempA=deepcopy_matrix(A)
    for i in range(0,day-1):
        tempA=multi_matrix(tempA,tempA)
        
    C=multi_matrix(tempA,v)
    return C

#Hàm check_positive_matrix: hàm kiểm tra ma trận A có chứa toàn phần tử >0 hay không
def check_positive_matrix(A):
    for i in range(len(A)):
        for j in range(len(A[i])):
            if A[i][j]<=0:
                return False
    return True

#Hàm check_regular_matrix: hàm kiểm tra ma trận A có chính quy hay không
def check_regular_matrix(A):
    C=multi_matrix(A,A)
    
    for i in range(0,50):
        if(check_positive_matrix(C)==True):
            return True
        C=multi_matrix(C,A)
    return False    

#Hàm steady_state_vec: hàm tìm vector phân phối dừng
def steady_state_vec(A1):
    
    if(check_regular_matrix(A1)==False):
        return None
    
    A=deepcopy_matrix(A1)
    
    for i in range(len(A)):
        A[i][i]=A[i][i]-1
    
    flag=True
    k=0
    j=0
    
    for i in range(0,len(A[0])):
        k=j
            
        flag=False
        
        for l in range(j,len(A)):
            if A[l][i] != 0:
                flag=True
                
        if flag == True:
            if A[k][i] == 0:
                for z in (k+1,len(A)):
                    if A[z][i] != 0:
                        temp=A[k]
                        A[k]=A[z]
                        A[z]=temp
                        break
            d=(1/A[k][i])
            for m in range(0,len(A[0])):
                A[k][m]=A[k][m]*d
            
            for k1 in range (0,len(A)):
                if k1==i:
                    continue
                d=-A[k1][i]/A[k][i]
                for m in range(0,len(A[0])):
                    A[k1][m]=round(A[k1][m]+d*A[k][m],3) 
            j=j+1
    V=[]
    for i in range(0,len(A)-1):
        V.append(A[i][len(A[i])-1])
        V[i]=-V[i]
    V.append(1)
    sum=0
    for i in range(0,len(V)):
        sum=sum+V[i]
    sum=round(1.0/sum,3)
    for i in range(0,len(V)):
        V[i]=round(V[i]*sum,3)   
        
    return V

#Hàm print_ans: hàm in kết quả test ra màn hình
def print_ans(A,v,day, state):
    
    print('[ DỰ BÁO THỜI TIẾT NGÀY HÔM SAU DỰA TRÊN THỜI TIẾT HIỆN TẠI ]')
    print_matrix(A,state)
    print('')
    print('Dựa vào ma trận chuyển trạng thái phía trên, có thể dự báo thời tiết ngày hôm sau như sau: ')
    for i in range(len(A[0])):
        print('')
        for j in range(len(A)):
            print('Nếu hôm nay '+state[i]+', tỉ lệ ngày mai '+state[j]+' là '+str(A[j][i]*100)+'%')
    
    print(' ')
    
    print('[ THỜI TIẾT HÔM NAY ]')
    for i in range(len(state)):
        print('Tỉ lệ '+state[i]+': ' + str(v[i][0]*100) + '%')    
    
    print(' ')
    
    for i in range(len(day)):
        vx=markov_chain(A,v,day[i])
        print('[ DỰ BÁO THỜI TIẾT SAU '+ str(day[i])+' NGÀY KỂ TỪ NGÀY HÔM NAY ]')
        for i in range(len(state)):
            print('Tỉ lệ '+state[i]+': ' + str(vx[i][0]*100) + '%') 
        print('')
        
    print('[DỰ BÁO THỜI TIẾT VỀ LÂU DÀI (PHÂN PHỐI DỪNG)]')
    V=steady_state_vec(A)
    if V != None:
        for i in range(len(state)):
            print('Về lâu dài, tỉ lệ '+state[i]+': ' + str(V[i]*100) + '%')
    else:
        print('Mô hình không có phân phối dừng!')
        
    print('')
    return

#----------------------------main-------------------------------------------
#data reference: https://earth-perspectives.springeropen.com/articles/10.1186/s40322-017-0042-6#:~:text=A%20Markov%20chain%20is%20a,dry%20on%20the%20previous%20day.
print('Họ tên: Đàm Tử Tâm')
print('MSSV: 21120551')
print('Lớp: 21_2')
print('')
print('DỰ BÁO THỜI TIẾT CÁC KHU VỰC TẠI GHANA')
print('')

print('')
print('------------------- THÀNH PHỐ CAPE COAST -------------------')
print('')

state=['mưa','nắng']
A=[[0.06,0.16],[0.94,0.84]]
v=[[1],[0]]
day=[1,2,3]

print_ans(A,v,day, state)
    

print('')
print('------------------- THỦ ĐÔ ACCRA -------------------')
print('')

state=['mưa','nắng']
A=[[0.10,0.13],[0.90,0.87]]
v=[[0],[1]]
day=[1,2,3]

print_ans(A,v,day, state)


print('')
print('------------------- THỊ TRẤN AKUSE -------------------')
print('')

state=['mưa','nắng']
A=[[0.13,0.06],[0.87,0.94]]
v=[[0],[1]]
day=[1,2,3]

print_ans(A,v,day, state)


print('')
print('------------------- THỊ TRẤN AKATSI -------------------')
print('')

state=['mưa','nắng']
A=[[0.10,0.06],[0.90,0.94]]
v=[[1],[0]]
day=[1,2,3]

print_ans(A,v,day, state)


print('')
print('------------------- THỊ TRẤN KETA -------------------')
print('')

state=['mưa','nắng']
A=[[0.03,0.06],[0.97,0.94]]
v=[[1],[0]]
day=[1,2,3]

print_ans(A,v,day, state)
```
