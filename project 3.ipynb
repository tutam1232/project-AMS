## Giải thuật Gram – Schmidt để phân rã ma trận A = QR
- Bước 1: giả sử ma trận $A$ có n cột, từng cột của ma trận sẽ là các vector từ {$u_1,u_2,...u_n$}
- Bước 2: đặt $v_1 = u_1$, nếu $v_1$ là vector 0 thì họ không độc lập tuyến tính và kết thúc giải thuật
- Bước 3: đặt $v_2,v_3,...v_n$ lần lượt với $v_m = u_m − \sum_{i=1}^{m-1}\frac{⟨u_m, v_i⟩}{∥v_i∥^{2}}v_i$ (m từ 2 đến n). Nếu có bất kì vector $v_m$ nào là vector 0 thì họ không độc lập tuyến tính và kết thúc giải thuật
- Bước 4: xây dựng họ các vector trực chuẩn với $q_m = \frac{v_m}{∥v_m∥}$ (m từ 1 đến n). 
- Bước 5: xây dựng ma trận $Q$ gồm n cột, $Q(q_1,q_2,...,q_n)$
- Bước 6: xây dựng ma trận R là một ma trận tam giác trên, với
R = $\begin{bmatrix} ⟨u_1, q_1⟩ & ⟨u_2, q_1⟩ & ... & ⟨u_n, q_1⟩ \\
    0 & ⟨u_2, q_2⟩ & ... & ⟨u_n, q_2⟩ \\
    \vdots &\vdots & \ddots &\vdots\\               
    0 & 0 & ... & ⟨u_n, q_n⟩ \end{bmatrix}$

## Mở rộng: tìm hiểu và trình bày ứng dụng của QR decomposition
Phân tích QR (QR decomposition) là một phương pháp phân tích ma trận trong đại số tuyến tính cho phép chia một ma trận bất kỳ thành tích của hai ma trận, một ma trận Q là một ma trận Orthogonal và một ma trận R là ma trận tam giác trên.

Phân tích QR là một công cụ rất hữu ích và được dùng phổ biến trong toán học, khoa học máy tính và các lĩnh vực khác. Một số ví dụ về ứng dụng của QR decomposition như sau:

- Giải hệ phương trình tuyến tính: Một trong những ứng dụng chính của phân tích QR là giải hệ phương trình tuyến tính. Vì ma trận R là ma trận tam giác trên, việc giải hệ phương trình sẽ dễ dàng hơn.

- Tìm trị riêng và vectơ riêng: Với phân tích QR, ta có thể tính toán trị riêng và vectơ riêng của một ma trận bằng cách sử dụng phương pháp QR algorithm.

- Tính toán đa thức: Phân tích QR có thể được sử dụng để tính toán đa thức của một ma trận.

- Giảm chiều dữ liệu: Phân tích QR cũng có thể được sử dụng để giảm chiều dữ liệu, trong đó ma trận Q có thể được sử dụng để ánh xạ dữ liệu vào một không gian chiều thấp hơn.

- Giải các bài toán tối ưu: Phân tích QR có thể được sử dụng để giải các bài toán tối ưu, bao gồm cả bài toán tối ưu tuyến tính và phi tuyến tính.

### Các hàm trong chương trình chính (không dùng thư viện):
**1. multi_matrix:** Hàm nhân hai ma trận, với input là ma trận A và B, nếu số cột của A khác số dòng của B thì kết thúc hàm (không thể nhân). Ngược lại, thực hiện 3 vòng for lồng nhau để nhân ma trận và trả về ma trận sau khi nhân A và B.\
**2. print_matrix:** Hàm in một ma trận, với input là ma trận cần in.\
**3. subtract_vector:** Hàm trừ hai vector, với input là vector A và vector B, ếu số phần tử của 2 vector khác nhau thì kết thúc hàm (không thể trừ). Ngược lại, thực hiện A - B và trả về vector sau khi thực hiện A - B.\
**4. multi_vector_const:** Hàm nhân một vector với một hằng số, với input là hằng số c và vector A, trả về vector sau khi thực hiện c * A\
**5. product:** Hàm tính tích vô hướng chính tắc của hai vector, với input là vector A và vector B, nếu số phần tử của hai vector khác nhau thì ngưng hàm, trả về 0. Ngược lại, trả về tích vô hướng chính tắc của hai vector.\
**6. is_0_vector:** Hàm kiểm tra một vector có là vector 0 không, nếu phải, trả về True, ngược lại trả về False\
**7. norm:** Hàm tính chuẩn của một vector, với input là vector A cần tính chuẩn, trả về chuẩn của vector A\
**8. print_conclusion:** Hàm in kết quả chạy test cho chương trình chính (không dùng thư viện), với input là ma trận A, hàm sẽ in ra ma trận A. Nếu có thể phân rã ma trận A thành ma trận Q, R bằng hàm qr_decomposition (hàm này dùng thuật giải Gram – Schmidt), hàm sẽ in ra cả ma trận Q, ma trận R và ma trận QxR (tính bằng hàm multi_matrix) để kiểm tra lại kết quả (QxR = A). Ngược lại, nếu không thể phân rã ma trận A thành ma trận Q, R bằng thuật giải Gram – Schmidt thì thông báo và ngưng hàm.\
**9. qr_decomposition:** Hàm phân rã QR ma trận A bằng giải thuật Gram – Schmidt, với input là ma trận A có n cột
- Đầu tiên, lưu n vector cột của A vào T, tương ứng là ${u_1, u_2,..., u_n}$ trong bước 1 của giải thuật
- Lưu $v_1 = u_1$, đưa vào Q
- Tính $v_2,v_3,...v_n$ lần lượt với $v_m = u_m − \sum_{i=1}^{m-1}\frac{⟨u_m, v_i⟩}{∥v_i∥^{2}}v_i$ (m từ 2 đến n). Chạy một vòng for i để duyệt từ T[1] đến T[n-1] tương ứng với $u_2,u_3,...u_n$. Với mỗi $u_m$ (T[i]), lưu $u_m$ vào temp2. Lồng thêm một vòng for j phía trong vòng for i ở trên để duyệt tương ứng từ $v_1, v_2,..., v_{m-1}$. Mỗi lần vòng for j chạy, tính $\frac{⟨u_m, v_j⟩}{∥v_j∥^{2}}v_j$ bằng cách kết hợp các hàm multi_vector_const, product, norm và kết hợp các phép toán rồi lưu vào temp, sau đó lấy temp2 - temp. Sau khi vòng for j chạy xong với mỗi i, ta được $v_m = u_m − \sum_{i=1}^{m-1}\frac{⟨u_m, v_i⟩}{∥v_i∥^{2}}v_i$. Kiểm tra nếu $v_m$ là vector 0 bằng hàm is_0_vector thì thông báo họ không độc lập tuyến tính và kết thúc giải thuật và trả về None, None. Ngược lại, đưa $v_m$ vào Q.
- Tính trực chuẩn của mỗi vector trực giao trong Q: Chạy một vòng for i duyệt qua từng vector trong Q và tính chuẩn của Q[i] bằng hàm norm rồi lưu vào chuan1. Chạy thêm một vòng for j lồng trong vòng for i duyệt qua từng phần tử của vector Q[i] và lấy Q[i][j]/chuan1.
- Tạo ma trận R: Chạy một vòng for i duyệt qua từng vector của Q (các vector trực chuẩn), tạo thêm một vòng for j lồng trong i duyệt qua từng vector của T (các vector cột của ma trận input), nếu j<i thì R[i][j] = 0, ngược lại R[i][j] sẽ là tích vô hướng chính tắc của Q[i] và T[j] (tính bằng hàm product). Sau bước này ta được ma trận R trong phân rã QR
- Tạo ma trận Q: Tạo Q1 lưu ma trận chuyển vị của ma trận Q hiện có, ta được ma trận Q1 là ma trận Q trong phân rã QR .
- Trả về ma trận Q1, R tương ứng ma trận Q, R sau khi phân rã QR ma trận A
